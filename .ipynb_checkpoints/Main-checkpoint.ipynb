# Plotting Player's Height's and Weight's Over Time

```python
df2.shape
```

```python
df1.head()
```

```python
df1.describe()
```

## Scatter Plot of Player's Heights

```python
plt.scatter(df1['born'],df1['height'])
plt.show()
```

## Scatter Plot of Player's Weights 

```python
plt.scatter(df1['born'], df1['weight'])
plt.show()
```

```python
df1 = df1[np.isfinite(df1['height'])]
sns.distplot(df1['height'])
plt.show()
```

```python
df1 = df1[np.isfinite(df1['weight'])]
sns.distplot(df1['weight'])
plt.show()
```

```python
plt.subplots(figsize=(30,25))
sns.barplot(y='born', x='height', data=df1, orient = 'h')
plt.show()
```

```python
plt.subplots(figsize=(30,25))
sns.barplot(y='born', x='weight', data=df1, orient = 'h')
plt.show()
```

 We find that the height of NBA players over time is more constant, whereas the weight of players increases over time 

# Now we are going to look at how 3 point shooting in the NBA has changed over time

```python
df2.columns
```

```python
df2 = df2[np.isfinite(df2['3PAr'])]
plt.scatter(df2['Year'], df2['3PAr'])
plt.show()
```

```python
plt.subplots(figsize=(30,25))
sns.barplot(x='Year', y='3PAr', data=df2, orient = 'v', hue='3PAr')
plt.show()
```

```python
plt.subplots(figsize=(30,25))
sns.barplot(x='Year', y='3PA', data=df2, orient = 'v', hue='3PA')
plt.show()
```
