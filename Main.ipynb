## Correlation Between Stats

```python

correlations = df2.corr()
mask = np.zeros_like(correlations, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
f, ax = plt.subplots(figsize=(20,15))
cmap = sns.diverging_palette(220, 10, as_cmap=True)
_ = sns.heatmap(correlations, mask=mask, cmap=cmap, vmax=.3, center=0,
                square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.show()
```

### This Heat Map Shows Similarities or Differences Between Different Stats
#### Ex. When a player's block percentage (BLK%) is high, their Free throw percentage is low (FT%). This makes sense as centers are usually better at blocking and worse at shooting free throws
